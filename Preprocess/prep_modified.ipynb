```python
import csv
import os
import pandas as pd

# Caminho para o diretório com os arquivos de entrada
diretorio_entrada = "D:/TCC-XAP-AWID/AWID-ATK-F-Trn"

# Caminho para o diretório onde o arquivo de saída será salvo
diretorio_saida = "D:/TCC-XAP-AWID"

# Caminho completo para o arquivo de saída
arquivo_saida_caminho = os.path.join(diretorio_saida, "saida.csv")

# Abrir o arquivo de saída uma única vez no diretório especificado
with open(arquivo_saida_caminho, "w", newline="") as arquivo_saida:
    writer = csv.writer(arquivo_saida)

    # Iterar por todos os arquivos no diretório especificado
    for arquivo in os.listdir(diretorio_entrada):
        caminho_completo = os.path.join(diretorio_entrada, arquivo)

        # Verificar se é um arquivo (e não um diretório)
        if os.path.isfile(caminho_completo):
            with open(caminho_completo, "r") as arquivo_entrada:
                for linha in arquivo_entrada:
                    writer.writerow(linha.strip().split(","))
```
